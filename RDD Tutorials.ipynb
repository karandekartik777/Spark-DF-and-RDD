```
from pyspark import SparkConf,SparkContext
```

```
conf=SparkConf().setAppName('ReadFile')
```

```
sc=SparkContext.getOrCreate(conf=conf)
```

```
text=sc.textFile('')
```

```
text.collect()
```

```
#use map function in spark RDD
# Map is used as a mapper of data from one state to other
# it will create a new RDD
# rdd.map(lambda.x:x.split())
```

```
rdd=sc.textFile('/FileStore/tables/sample_txt-1.txt')
```

```
rdd2=rdd.map(lambda x:x.split(' '))
rdd2.collect()
```

```
def splitx(x):
  return x.split(' ')


rdd3=rdd.map(splitx)
rdd3.collect()
```

```
 new_rdd=sc.textFile('/FileStore/tables/sample1_txt-1.txt')
```

```
new_rdd.collect()
```

```
def quiz(x):
  l=x.split(' ')
  l3=[]
  for s in l:
    l3.append(len(s))
  return l3

rdd2=new_rdd.map(quiz)
rdd2.collect()
```

```
# flatMap()
# flat Map is used as a maper of data and explodes data before final output
# it will create a new RDD
# rdd.flatMap(lambda x:x.split())
```

```
flatmappedrdd=rdd.flatMap(lambda x:x.split(''))
flatmappedrdd.collect()
```
