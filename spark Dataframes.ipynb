```
from pyspark import SparkConf,SparkContext
conf=SparkConf().setAppName('RDD')
sc=SparkContext.getOrCreate(conf=conf)

rdd=sc.textFile('/FileStore/tables/StudentData.csv')
headers=rdd.first()
rdd=rdd.filter(lambda x:x!=headers).map(lambda x :x.split(','))
```

```
# converting RDD to Dataframe

columns=headers.split(',')
dfrdd=rdd.toDF(columns)
dfrdd.show()
dfrdd.printSchema()
```

```
columns=headers.split(',')
```

```
# U can also change the dtype of the schema
dfrdd=spark.createDataFrame(rdd,schema=schema)
dfrdd.printSchema()
```

```
#when you are usinggroupny on multiple datacolumns then you want to perform operations like agg sum avg they how can you do that we can see it the steps below


#------->>>>for that you have to import aggregative function
from pyspark.sql.functions import sum, avg,min,max ,mean, count

#by importing above function you can apply multiple operations
df.groupBy('course').agg(count('*'),sum('marks'),min('marks'),max('marks'),avg('marks')).show()
```

```
 #if you say the above the name of newly created columns is so weird so we can change the name to the new names using alias
  
df.groupBy('course').agg(count('*').alias('total_enrollments'),sum('marks').alias('total_marks'),min('marks').alias('lowest_marks'),max('marks').alias('highest_marks'),avg('marks')).show()
```
